```python
# setup imports
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
```

```python
# setup browser
options = Options()
options.headless = True
b = webdriver.Chrome(options=options)
```

```python
# scrape webpage
url = "https://docs.google.com/spreadsheets/u/1/d/e/2PACX-1vTwkB_zRc1yCzyMS8QMKo6mBPOsLQGycnhKRLK7w7_duWrwSv65UNKwNcB-YduYWsL8QRysv59cZnIQ/pubhtml?gid=1169843611&single=false#"
b.get(url)
```

```python
# convert webpage to CSV
import pandas as pd
df = pd.read_html(b.page_source)[1]
df.to_csv("raw_march_randomness.csv")
```
