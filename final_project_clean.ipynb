```python
# libraries
import pandas as pd
import numpy as np
# import re
```

```python
# read CSV into DataFrame, check again
df_raw = pd.read_csv("raw_march_randomness.csv")
df_raw.head(10)
```

```python
df = df_raw.iloc[3:, 2:]
df = df.reset_index(drop=True)
df.columns = ["rank", "team_name", "final_StatCoin_balance", "bet_each_round", "edu_level", "competitor_type",
              "day_1_guess", "day_1_bet", "day_2_guess", "day_2_bet", "day_3_guess", "day_3_bet", "day_4_guess", "day_4_bet",
              "day_5_guess", "day_5_bet", "day_6_guess", "day_6_bet", "day_7_guess", "day_7_bet", "day_8_guess", "day_8_bet", 
              "day_9_guess", "day_9_bet", "day_10_guess", "day_10_bet", "day_11_guess", "day_11_bet", "day_12_guess", "day_12_bet",
              "day_13_guess", "day_13_bet", "day_14_guess", "day_14_bet", "day_15_guess", "day_15_bet", "day_16_guess", "day_16_bet"]

# noticed weird problem, value missing under "bet_each_round"
# so I checked to see if there were multiple like this
    # df[(df["bet_each_round"] != 'yes') & (df["bet_each_round"] != 'no')]
# thankfully only one was missing, so I filled it in logically:
df.at[2,"bet_each_round"] = 'yes'
df.head(10)
# so far so good
```

```python
# make sure numbers aren't strings

df["rank"] = df["rank"].astype(int)

# I didn't know about the .replace() method for DataFrames!
# code adapted from: https://stackoverflow.com/questions/32464280/converting-currency-with-to-numbers-in-python-pandas
df["final_StatCoin_balance"] = df["final_StatCoin_balance"].replace('[/$,]','', regex=True).astype(int)

# convert all bets to floats (including Free Bets)
for column in ["day_1_bet", "day_2_bet", "day_3_bet", "day_4_bet",
               "day_5_bet", "day_6_bet", "day_7_bet", "day_8_bet",
               "day_9_bet", "day_10_bet", "day_11_bet", "day_12_bet",
               "day_13_bet", "day_14_bet", "day_15_bet", "day_16_bet"]:
    df[column] = df[column].replace('Free Bet', '1500', regex=True)
    df[column] = df[column].astype(float)
df
```

```python
# at this point I noticed the `day_14_guess` column had a `yes` in it instead of a `Yes`
# so I decided to check all string columns to make sure they made sense

# quick check to make sure all team_names are strings:
for name in list(df["team_name"].values):
    assert type(name) == str, "that's not a string"
# thankfully all strings

# now I actually care that each "edu_level" is specifically "Pre-College" or "College"
# so let's check that:
for level in list(df["edu_level"].values):
    assert (level == "Pre-College") | (level == "College"), "that's not an edu level"
# excellent, that's good

# same for competitor type, either "Team" or "Individual":
for comp_type in list(df["competitor_type"].values):
    assert (comp_type == "Team") | (comp_type == "Individual"), "that's not a competitor type"
# that's all good


# now on to the guesses.  First I'll change the one random "yes" I saw:

df.at[0, "day_14_guess"] = "Yes"

# now let's check to make sure everything is either "Yes", "No" or NaN for each guess
for guess_day in ["day_1_guess", "day_2_guess", "day_3_guess", "day_4_guess",
                  "day_5_guess", "day_6_guess", "day_7_guess", "day_8_guess",
                  "day_9_guess", "day_10_guess", "day_11_guess", "day_12_guess",
                  "day_13_guess", "day_14_guess", "day_15_guess", "day_16_guess"]:
    for guess in list(df[guess_day].values):
        if guess not in {"Yes", "No"}:
            assert np.isnan(guess), "the guess is not Yes, No, or NaN"
# excellent! Turns out it was just that one typo

df
```

```python
df.to_csv("clean_march_randomness.csv")
```
